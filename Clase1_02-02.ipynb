```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 20) # Vector de tiempo
signal = np.sin(t)              # Señal analogica
```

```python
# Grafica de la señal analógica y digital
plt.figure()
plt.plot(t, signal, 'b', label = 'Señal analógica')
plt.stem(t, signal, 'r', label = 'Señal digitalizada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```

## Ejemplos propios:
### Digitalización y reconstrucción de una señal analógica.

### Ejemplo 1
Este ejemplo crea una señal sinusoidal analógica y la digitaliza tomando muestras a una frecuencia de muestreo de 100 Hz. Luego, se reconstruye la señal digitalizada a su forma original mediante la interpolación lineal. Finalmente, se grafican ambas señales para compararlas.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar señal analógica
t = np.linspace(0, 1, 1000)
analog_signal = np.sin(2 * np.pi * t)

# Digitalizar señal
sampling_frequency = 100
sampled_signal = analog_signal[::int(1 / (1 / sampling_frequency))]

# Reconstruir señal
t_sampled = np.linspace(0, 1, len(sampled_signal))
reconstructed_signal = np.interp(t, t_sampled, sampled_signal)

# Mostrar señal analógica y reconstruida
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t, analog_signal, label='Analog Signal', linewidth=3)
ax.plot(t, reconstructed_signal, label='Reconstructed Signal', linewidth=3)
ax.scatter(t_sampled, sampled_signal, label='Sampled Signal', color='red')
ax.set_xlabel('Time (s)', fontsize=12)
ax.set_ylabel('Signal Value', fontsize=12)
ax.legend(fontsize=12)
plt.tight_layout()
plt.grid()
plt.show()
```

### Ejercicio 2:
Este ejemplo crea una señal sinusoidal analógica y la digitaliza utilizando un umbral. Todos los valores de la señal analógica mayores que 0.5 se convierten en 1, y todos los valores menores o iguales a 0.5 se convierten en -1. Finalmente, se grafican ambas señales para compararlas. La señal digitalizada se representa mediante una gráfica de escalones.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar señal analógica
t = np.linspace(0, 1, 1000)
analog_signal = np.sin(2 * np.pi * 10 * t)

# Digitalizar señal
sampling_frequency = 50
sampled_signal = np.zeros(len(t))
threshold = 0.5
sampled_signal[analog_signal > threshold] = 1
sampled_signal[analog_signal <= threshold] = -1

# Mostrar señal analógica y digitalizada
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t, analog_signal, label='Analog Signal', linewidth=3)
ax.step(t, sampled_signal, label='Digital Signal', where='post', linewidth=3)
ax.set_xlabel('Time (s)', fontsize=16)
ax.set_ylabel('Signal Value', fontsize=16)
ax.legend(fontsize=16)
plt.tight_layout()
plt.grid()
plt.show()
```
